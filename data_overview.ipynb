# Data Overview

This notebook provides a **deep statistical overview** of every CSV file in the `Dataset/` directory.  
For each dataset we display:

| Section | What it shows |
|:--------|:--------------|
| **First 5 Rows** | Preview of the actual data using `.head(5)` |
| **Info** | Column names, non-null counts, dtypes, and memory usage |
| **Describe (Numeric)** | Count, mean, std, min, 25%, 50%, 75%, max for numeric columns |
| **Describe (Categorical)** | Count, unique, top value, frequency for non-numeric columns |
| **Missing Values** | Columns with NaN counts and percentage |

````python
# ──────────────────────────────────────────────
# Imports & Display Settings
# ──────────────────────────────────────────────

import io
import pandas as pd
from pathlib import Path
from IPython.display import display, Markdown, HTML

# Show all columns/rows in describe output (no truncation)
pd.set_option("display.max_columns", None)
pd.set_option("display.max_rows", None)
pd.set_option("display.width", None)
pd.set_option("display.float_format", "{:,.2f}".format)
````

````python
# ──────────────────────────────────────────────
# Configuration
# ──────────────────────────────────────────────

DATASET_DIR = Path("Dataset")
csv_files = sorted(DATASET_DIR.glob("*.csv"))

print(f"Found {len(csv_files)} CSV file(s) in '{DATASET_DIR}/'\n")
for i, f in enumerate(csv_files, 1):
    print(f"  {i:>2}. {f.name}")
````

````python
# ──────────────────────────────────────────────
# Helper: capture df.info() as a string
# ──────────────────────────────────────────────

def get_info_string(df: pd.DataFrame) -> str:
    """
    Capture the output of df.info() into a string
    since info() prints directly to stdout.
    """
    buf = io.StringIO()
    df.info(buf=buf, show_counts=True, memory_usage="deep")
    return buf.getvalue()
````

````python
# ──────────────────────────────────────────────
# Helper: build a missing-values summary table
# ──────────────────────────────────────────────

def missing_summary(df: pd.DataFrame) -> pd.DataFrame:
    """
    Return a DataFrame showing only columns that have
    missing values, with count and percentage.
    """
    missing = df.isnull().sum()
    missing = missing[missing > 0].sort_values(ascending=False)

    if missing.empty:
        return None

    summary = pd.DataFrame({
        "Missing Count": missing,
        "% of Total": (missing / len(df) * 100).round(2)
    })
    summary.index.name = "Column"
    return summary
````

````python
# ──────────────────────────────────────────────
# Main: iterate through every CSV
# ──────────────────────────────────────────────

for idx, csv_path in enumerate(csv_files, 1):

    # Read the file
    df = pd.read_csv(csv_path)
    rows, cols = df.shape

    # ── Header ──────────────────────────────────
    display(Markdown(f"---"))
    display(Markdown(
        f"## {idx}. `{csv_path.name}`\n\n"
        f"**Rows:** `{rows:,}` &nbsp;|&nbsp; "
        f"**Columns:** `{cols}`"
    ))

    # ── 1. First 5 Rows ────────────────────────
    display(Markdown("### 🔍 First 5 Rows"))
    display(df.head(5))

    # ── 2. df.info() ────────────────────────────
    display(Markdown("### 📋 Data Info"))
    info_str = get_info_string(df)
    display(Markdown(f"```\n{info_str}\n```"))

    # ── 3. df.describe() — Numeric columns ──────
    numeric_desc = df.describe()
    if not numeric_desc.empty and numeric_desc.shape[1] > 0:
        display(Markdown("### 📊 Describe — Numeric Columns"))
        display(numeric_desc)

    # ── 4. df.describe() — Categorical columns ──
    cat_desc = df.describe(include=["object"])
    if not cat_desc.empty and cat_desc.shape[1] > 0:
        display(Markdown("### 🔤 Describe — Categorical Columns"))
        display(cat_desc)

    # ── 5. Missing values summary ───────────────
    ms = missing_summary(df)
    if ms is not None:
        display(Markdown("### ⚠️ Missing Values"))
        display(ms)
    else:
        display(Markdown("### ✅ No Missing Values"))

    print()  # spacing between datasets
````
